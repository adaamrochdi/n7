```python
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d
import numpy as np
import scipy.sparse as scps
import scipy.sparse.linalg as ssl
import scipy.linalg as sl
import math
from utils import *
```

**Partie I : maillage triangulaire et conditions de Dirichlet**

```python
n = 20
coord, triangles, dirichlet, neumann = maillage(n)
Nt = np.size(triangles, 0)
Ns = np.size(coord, 0)

A = np.zeros((Ns,Ns))
b = np.zeros((Ns,1))
U = np.zeros((Ns,1))

for s in range(Ns):
    if s in dirichlet:
        U[s] = u_d(coord[s][0],coord[s][1])
    else:
        U[s] = 0

for p in range(Nt):
    i, j, l = triangles[p]
    
    x = np.array([coord[i][0], coord[j][0], coord[l][0]])
    y = np.array([coord[i][1], coord[j][1], coord[l][1]])
    
    x_G = np.mean(x)
    y_G = np.mean(y)
    
    M = calcul_M(x, y)
    
    for m in range(3):
        for k in range(3):
            A[triangles[p][m]][triangles[p][k]] += M[m][k]

        b[triangles[p][m]] += alpha(x,y) / 6 * f(x_G,y_G)

b -= np.matmul(A,U)

indice = np.setdiff1d([s for s in range(Ns)], dirichlet)

Uh = U


As = np.zeros((len(indice),len(indice)))
bs = np.zeros((len(indice),1))

for i in range(len(indice)):
    for j in range(len(indice)):
        As[i][j] = A[indice[i]][indice[j]]
    
    bs[i] = b[indice[i]]
        
z = sl.solve(As, bs)

Uh[indice] = z

show(coord, Uh[:,0])
```

**Partie II : maillage mixte et ajot des conditions de Neumann**

**Calcul de la matrice de raideur**

La matrice des raideur M pour les quadrangles est donne par : $[M_T^A]_{ij} = |J| \iint_{T} \nabla \phi_i ^\top (s,t) \ (J_TJ_T^\top)^{-1} \ \nabla \phi_j (s,t) \ dsdt$.


En posant $(J_TJ_T^\top)^{-1} = \begin{bmatrix}
a & b \\
b & c \\
\end{bmatrix}$ et en faisant le calcul des 10 $m_{ij}$, on retrouve :

$M_T^A = \begin{bmatrix}
2a + 3b + 2c & -2a + c & -a -3b -c & a - 2c \\
-2a + c & 2a - 3b + 2c & a - 2c & -a + 3b -c \\
-a -3b -c & a - 2c & 2a + 3b + 2c & -2a + c \\
a - 2c & -a + 3b -c & -2a + c & 2a - 3b + 2c \\
\end{bmatrix}$

```python
############################# Maillage mixte ################
e3=np.array([[1,2,12],[2,3,12],[3,4,14],[4,5,14],[2,15,3],[3,15,4]]).astype(int)
e4=np.array([[0,1,12,11],[11,12,13,10],[12,3,14,13],[10,13,8,9],[13,14,7,8],[14,5,6,7]]).astype(int)
dds=np.array([2,15,4,6,7,8,9,10,11,0]).astype(int)
nns=np.array([[4,5],[5,6],[0,1],[1,2]]).astype(int)
ccs=np.array([[0.,0.],[0.33333333333333,0],[0.53333333333333,0.],
                      [0.66666666666667,0.33333333333333],[1.,0.47],[1,0.66666666666667],
                     [1.,1.],[0.66666666666667,1.],[0.33333333333333,1.], [0.,1.],
                     [0.,0.66666666666667],[0.,0.33333333333333],[0.33333333333333,0.33333333333333],
                     [0.33333333333333,0.66666666666667],[0.66666666666667,0.66666666666667],[1.,0.]])
```

```python
Nt = np.size(e3, 0)
Nq = np.size(e4, 0)
Ns = np.size(ccs, 0)
Nn = np.size(nns, 0)

A = np.zeros((Ns,Ns))
b = np.zeros((Ns,1))
U = np.zeros((Ns,1))

# boucle triangles
for p in range(Nt):
    i, j, l = e3[p]
    
    x = np.array([ccs[i][0], ccs[j][0], ccs[l][0]])
    y = np.array([ccs[i][1], ccs[j][1], ccs[l][1]])
    
    x_G = np.mean(x)
    y_G = np.mean(y)
    
    Mt = calcul_M(x, y)
    
    for m in range(3):
        for k in range(3):
            A[e3[p][m]][e3[p][k]] += Mt[m][k]

        b[e3[p][m]] += alpha(x,y) / 6 * f(x_G,y_G)

# boucle sur quadrangles
for q in range(Nq):
    i, j, k, l = e4[q]
    
    x = np.array([ccs[i][0], ccs[j][0], ccs[k][0], ccs[l][0]])
    y = np.array([ccs[i][1], ccs[j][1], ccs[k][0], ccs[l][1]])
    
    Mq = calcul_M_quad(x,y)
    
    for m in range(4):
        for t in range(4):
            A[e4[q][m]][e4[q][t]] += Mq[m][t]

        b[e4[q][m]] += alpha(x,y) / 4 * f(x_G,y_G)
        
for s in range(Ns):
    if s in dds:
        U[s] = u_d(ccs[s][0],ccs[s][1])
    else:
        U[s] = 0

# Conditions de Dirichlet non-homognes       
b -= np.matmul(A,U)

# Conditions de Neumann
# Boucle sur arrtes Neumann
for p in range(Nn):
    i, j = nns[p]
    x_G = (ccs[i][0] + ccs[j][0]) / 2
    y_G = (ccs[i][1] + ccs[j][1]) / 2
    integrale = math.dist(ccs[i], ccs[j]) * g(x_G, y_G) / 2
    b[i] += integrale
    b[j] += integrale


# Recherche des sommets en dehors de la frontire de Dirichlet
indice = np.setdiff1d([s for s in range(Ns)], dds)

# Solution discrte
Uh = U

As = np.zeros((len(indice),len(indice)))
bs = np.zeros((len(indice),1))


for i in range(len(indice)):
    for j in range(len(indice)):
        As[i][j] = A[indice[i]][indice[j]]
    
    bs[i] = b[indice[i]]
        
        
z = sl.solve(As, bs)

Uh[indice] = z

show(ccs, Uh[:,0])
```

**Complments  :  un nouweau terme dans l'EDP**

**La formulation wariationnelle**


Soit $(u,w) \in w^2 = H_0^1(\Omega)^2$, on a :

$\int_{\Omega}-\Delta uwdx + \int_{\Omega}c_0uwdx = \int_{\Omega}fw dx$

De plus, $\Omega$ ouvert born  frontires lipschitzienne donc d'aprs la formule de Green :

$\int_{\Omega}\Delta uwdx = -\int_{\Omega}\nabla u \cdot \nabla w dx + \int_{\Gamma} \gamma_{0}(w) \cdot \gamma_1(u) d\gamma$

Donc la formulation variationelle du problme peut s'crire :

Trouver $u \in$ V telle que 


$\forall w \in H_0^1(\Omega)$,

$\int_{\Omega}\nabla u \cdot \nabla w dx - \int_{\Gamma} \gamma_{0}(w) \cdot \gamma_1(u) d\gamma + \int_{\Omega}c_0uwdx = \int_{\Omega}fw dx $

$\Leftrightarrow  \int_{\Omega}\nabla u \cdot \nabla w dx + \int_{\Omega}c_0uwdx = \int_{\Omega}fw dx$ (car $\gamma_1(u) = 0$)

**Unicit de la solution**

On note $a(u,w) = \int_{\Omega}\nabla u \cdot \nabla w dx + c_0\int_{\Omega}uwdx$ et $l(w) = \int_{\Omega}fwdx$

1. $H_0^1(\Omega)$ est espace de Hilbert
2.

- a est bilinaire par linarit de l'intgrale
- Soit $(u,w) \in V^2$, on a :

$|a(u,w)| \leq |\int_{\Omega}\nabla u \cdot \nabla w dx| + c_0|\int_{\Omega}uwdx|$

$\leq   <u,w>_{1,\Omega} + <u,w>_{L^2(\Omega)}$

$\leq |u|_{1, \Omega}|w|_{1, \Omega} + c_0||u||_{L^2(\Omega)}||w||_{L^2(\Omega)}$

$\leq (1+c_0)||u||_{H^1(\Omega)}||w||_{H^1(\Omega)}$

$(1+c_0)$ est une constante strictement positive donc a est continue.

- Soit $w \in V$, on a :
$a(w,w) = |w|_{1, \Omega}^2 + c_0||w||_{L^2(\Omega)}^2 $

$\geq |w|_{1, \Omega}^2$

$\geq ||w||_{H^1(\Omega)}^2$

D'o a est coercive.

3. - l est linaire par linarit de l'intgrale
   
Soit w $\in V^2$, on a :

$|l(w)| = |\int_{\Omega}fwdx|$

$\leq ||f||_{L^2(\Omega)}||w||_{L^2(\Omega)}$

$\leq ||f||_{H^1(\Omega)}||w||_{H^1(\Omega)}$

Or $||f||_{H^1(\Omega)} > 0$ donc l est continue. 



D'aprs le thorme de Lax-Milgram, il existe une unique solution.

**Systme linaire d'quations $Ax=b$**

Soit $V_h$ un sous-espace vectoriel de $V$ tel que $N_{h} = \operatorname{dim} (V_h) < + \infty$ avec $h \sim \frac{1}{N_{s}}$ et $(v_{i} )\in V_{h}^{N_{h}}$ une base de $V_h$

$$
\begin{gather}
\exists ! (\alpha_{i}) \in \mathbb{R}^{N_{h}}, \ u_{h} = \sum_{i=1}^{n} \alpha_{i} w_{i} \\
u_{h} \text{ solution de } (P_{FV_{h}}) \Leftrightarrow \forall v_{h} \in V_{h}, \ a(u_{h},v_{h}) = l(v_{h}) \\
\Leftrightarrow \forall i \in [\![1,N_{h}]\!], \ a(u_{h},w_{i}) = l(w_{i}) \\
\Leftrightarrow \forall i \in [\![1,N_{h}]\!], \ \sum_{j=1}^{N_{h}} \alpha_{j} a(w_{j},w_{i}) = l(w_{i}) \text{ par limite  gauche de a} \\
\Leftrightarrow A \alpha = b \text{ avec } A \in \mathcal{M}_{N_{h}} (\mathbb{R}) \text{ telle que } A_{ij} = a(w_{j},w_{i}), \ \forall (i,j) \in [\![1,N_{h}]\!]^{2} \\ b \in \mathbb{R}^{N_{h}} \text{ tel que } b = l(w_{i}) \ \forall i \in [\![1,N_{h}]\!] \text{ et } \alpha = (\alpha_{i})_{i \in \mathbb{R}^{N_{h}}}
\end{gather}
$$

D'aprs le lemme de Cea : $\exists !\ u_h \in V_h, \forall v_h \in V_h : a(u_h,v_h) = l(v_h)$




```python
rcn_n = 10
n = rcn_n**2
coord, triangles, dirichlet, neumann = maillage(rcn_n)
Nt = np.size(triangles, 0)
Ns = np.size(coord, 0)
c0 = 5

A = np.zeros((Ns,Ns))
b = np.zeros((Ns,1))
U = np.zeros((Ns,1))

for p in range(Nt):
    i, j, l = triangles[p]
    
    x = np.array([coord[i][0], coord[j][0], coord[l][0]])
    y = np.array([coord[i][1], coord[j][1], coord[l][1]])
    
    x_G = np.mean(x)
    y_G = np.mean(y)
    
    alpha = alpha(x,y)
    
    M = calcul_M(x, y)
    
    
    for m in range(3):
        for k in range(3):
            A[triangles[p][m]][triangles[p][k]] += M[m][k]

        b[triangles[p][m]] += alpha / 6 * f(x_G,y_G)

b -= np.matmul(A,U)

indice = np.setdiff1d([s for s in range(Ns)], dirichlet)

Uh = U

As = np.zeros((len(indice),len(indice)))
bs = np.zeros((len(indice),1))

for i in range(len(indice)):
    for j in range(len(indice)):
        As[i][j] = A[indice[i]][indice[j]]
    
    bs[i] = b[indice[i]]
        
z = sl.solve(As, bs)

Uh[indice] = z

show(coord, Uh[:,0])
```
